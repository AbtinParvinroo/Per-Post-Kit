```python
import numpy as np
import pandas as pd

class FeatureStatistics:
    def __init__(self, df):
        self.df = df

    def variance(self, column_name:str):
        self.variance_rate = np.var(self.df[column_name])
        return self.variance_rate

    def standard_divition(self, column_name:str):
        self.std_rate = np.std(self.df[column_name])
        return self.std_rate

    def correlation(self, column_name:str, method:str='pearson'):
        self.correlation_rate = self.df.corr(method=method)
        return self.correlation_rate
```

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

class feature_engineering_system:
    def __init__(self, dataset: pd.DataFrame):
        self.dataset = dataset.copy()

    def polynomial_features(self, config: dict):
        features = config.get('features', self.dataset.columns.tolist())
        degree = config.get('degree', 2)
        include_bias = config.get('include_bias', False)
        interaction_only = config.get('interaction_only', False)

        poly = PolynomialFeatures(degree=degree, include_bias=include_bias, interaction_only=interaction_only)
        poly_data = poly.fit_transform(self.dataset[features])
        poly_columns = poly.get_feature_names_out(features)
        poly_df = pd.DataFrame(poly_data, columns=poly_columns, index=self.dataset.index)

        self.dataset = pd.concat([self.dataset, poly_df], axis=1)
        return self.dataset

    def linear_features(self, config: dict):
        features = config.get('features', self.dataset.columns.tolist())
        scale = config.get('scale', 1.0)

        for f in features:
            new_col = f"{f}_linear_scaled"
            self.dataset[new_col] = self.dataset[f] * scale

        return self.dataset
```
